```python
#extracting the sequence from the fasta file
covid_nucleotide_sequence = COVID19_Sequence.seq
covid_nucleotide_sequence
```

```python
#printing the length of the sequence
print("The length of the COVID19 Sequence is::", len(covid_nucleotide_sequence))
```

```python
#creating a custom function to calculate the GC content of the sequene
def GC_Content(seq):
    result = float(seq.count('G') + seq.count('C'))/len(seq) * 100
    return result

print("The GC content of the COVID19 sequence is ---->", GC_Content(covid_nucleotide_sequence))
```

```python
#translating the sequence into a protein sequence
covid_protein_sequence = covid_nucleotide_sequence.translate()
print("The translated sequence of COVID19 is----->", covid_protein_sequence)
```
