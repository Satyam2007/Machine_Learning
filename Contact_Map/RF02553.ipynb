```python
#importing the packages 
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
#importing the csv file containing the contacts of the i,j pairs using pandas
#Rfam - RF00011 - Y RNA-like

df1 = pd.read_csv("RF02553.csv")
df1.head()
```
