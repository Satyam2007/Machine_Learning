```python
#importing the packages 
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
#importing the csv file containing the contacts of the i,j pairs using pandas
#Rfam - RF00011 - Bacterial RNase P class B
df1 = pd.read_csv("RF02266.csv")
df1.head()
```

```python
df1_heatmap = pd.pivot_table(df1, values='z', index=['y'], columns='x')
plt.figure(figsize=(12,10))
sns.heatmap(df1_heatmap, vmin=-0.2, vmax=0.2)
plt.show()
```
